## Custom Query Completion

TODO: In the cells below, compose a custom query using your chosen dataset and retrieve results from an OpenAI `Completion` model. You may copy and paste any useful code from the course materials.

```python
import tiktoken
tokenizer = tiktoken.get_encoding("cl100k_base")

prompt_template = '''
Question:{}
Answer:
'''
```

```python
def custom_sys(df, question, model="gpt-3.5-turbo-instruct", max_tokens=2900):
    # Initialize the tokenizer
    tokenizer = tiktoken.get_encoding("cl100k_base")  # Assuming this is the encoding you need
    
    custom_prompt = '''You are a product expert. Answer the question based on the product review provided in the context. 
If the question cannot be answered based on the review, say "I don't know."
Provide detailed and informative answers. If possible, mention key features, pros, and cons of the product.
Context: {}
- - - - - -
Question: {}
Answer:
'''

    custom_counter = 0
    custom_list = []

    # Loop through the reviews in the DataFrame
    for answer in df['text'].values:
        current_custom_counter = len(tokenizer.encode(answer))
        custom_counter += current_custom_counter

        # Stop if the total token count exceeds the max_tokens limit
        if custom_counter > max_tokens:
            break
        
        custom_list.append(answer)
    
    # Format the prompt with the gathered context and question
    formatted_prompt = custom_prompt.format(
        " ".join(custom_list),
        question
    )

    try:
        # Request to the OpenAI API
        custom_response = openai.Completion.create(
            model=model,
            prompt=formatted_prompt,
            max_tokens=1000  # Adjust max tokens for the response
        )
        
        # Extract and return the answer
        custom_answer = custom_response["choices"][0]["text"].strip()
        return custom_answer
    
    except openai.error.OpenAIError as e:
        # Handle any API errors
        return f"Error with the OpenAI API: {str(e)}"
    except Exception as e:
        # General error handling
        return f"An error occurred: {str(e)}"
```

```python
user_question1 = "What are the main fashion trends for Spring 2023?"
user_question2 = "Can you describe the key details about the 'Sheer Clothing' trend for 2023?"
user_question3 = "What are some ways to style the 'Shiny Fabrics' trend in 2023?"
user_question4 = "What makes the 2023 'Cobalt Blue' trend stand out from other color trends?"
```
